Question 1 : find the area of a rectangle





```python
length = int(input("Enter length: "))
breadth = int(input("Enter breadth: "))

area = length * breadth

print("Area of rectangle is", area)
```

Queston 2 : Find simple interest

```python
principal_amount = int(input("enter principal amount :"))
rate = int(input("enter rate :"))
time = int(input("enter time :"))
simple_interest = (principal_amount * rate * time)/100
print("simple interest is",simple_interest)
```

Questioon 3 : Convert temperature from celsius to fahrenheit

```python
temperature_in_celsius = int(input("enter temperature in celsius :"))
temperature_in_fahrenheit = (temperature_in_celsius * 9/5) + 32
print("temperature in fahrenheit is",temperature_in_fahrenheit)
```

Question 5 : Find square and cube of a number

Question 7 :Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage , Use comments, Use proper indentations

```python
#taking input
name = input("enter name :")
roll_number = input("enter roll number :")
math_marks = int(input("enter math marks :"))
science_marks = int(input("enter science marks :"))
english_marks = int(input("enter english marks :"))
sst_marks = int(input("enter sst marks :"))
hindi_marks = int(input("enter hindi marks :"))
total_marks = math_marks + science_marks + english_marks + sst_marks + hindi_marks
percentage = (total_marks / 500) * 100
#printing output as report
print(name,"(",roll_number,")")
print("math marks :",math_marks)
print("science marks :",science_marks)
print("english marks :",english_marks)
print("sst marks :",sst_marks)
print("hindi marks :",hindi_marks)
print("total marks :",total_marks)
print("percentage :",percentage)
```

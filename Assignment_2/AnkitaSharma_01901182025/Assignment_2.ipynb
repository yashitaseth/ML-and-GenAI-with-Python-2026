```python
#1. Sum of first 10 natural numbers
sm=0
for i in range(1,11):
  sm+=i #adding all numbers to sum one by one
print("Sum of first 10 natural numbers is: ",sm) #printing the output
```

```python
#2. Find factorial of a number
num= int(input("Enter the number whose factorial is to be found:"))
fact=1
for i in range(1,num+1):
  fact*=i
print("Factorial of",num,'is: ',fact)
```

```python
#3. Print Fibonacci Series
Fibo= [1,1]
length= int(input("Enter the number of terms you want to print in the Fibonacci series:"))
for i in range(length-2):
  Fibo.append(Fibo[-1]+Fibo[-2])
print("Fibonacci series is:", Fibo)
```

```python
#5. Create Student Result system and display grade
name= input("Enter student name:")
marks= input("Enter student marks in dictionary: ") #taking input of all the marks, subject-wise
marks_dict = eval(marks) #converting the marks into dictionary
marks_list=marks_dict.values()
total=0
for i in marks_list:
  total+=i
percent= total/len(marks_list)
print("Percentage obtained by",name,"is:",percent)
if percent>=93 and percent<=100:
  print("Grade obtained by",name,"is: A+")
elif percent>=85 and percent<93:
  print("Grade obtained by",name,"is: A")
elif percent>=77 and percent<85:
  print("Grade obtained by",name,"is: B+")
elif percent>=69 and percent<77:
  print("Grade obtained by",name,"is: B")
elif percent>=61 and percent<69:
  print("Grade obtained by",name,"is: C")
elif percent>=53 and percent<61:
  print("Grade obtained by",name,"is: C+")
elif percent>=45 and percent<53:
  print("Grade obtained by",name,"is: D")
else:
  print("Grade obtained by",name,"is: F")
```

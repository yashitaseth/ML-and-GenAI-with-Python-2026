```python
#2 factorial
n=int(input("Enter a number:"))
factorial=1
for i in range(1,n+1):
  factorial*=i
print(factorial)
```

```python
#5 Student Result System
name=str(input("Enter student name:"))
roll=int(input("Enter roll no:"))
subjects = ["Maths", "Science", "Social Science", "English", "Hindi"]
total = 0
for i in subjects:
    marks = int(input("Enter marks in " + i + ": "))
    while marks < 0 or marks > 100:
        marks = int(input("Invalid input: "))
    total += marks
percentage=(total/500)*100
print("Total marks obtained:",total,"out of 500.")
print("Percentage marks secured:",percentage,"%")
if percentage >= 93:
  grade="A+"
elif percentage >= 85:
  grade="A"
elif percentage >=77:
  grade="B+"
elif percentage >=69:
  grade="B"
elif percentage >=61:
  grade="C+"
elif percentage >=53:
  grade="C"
elif percentage >=54:
  grade="D"
else:
  grade="F(Fail)"
print("Grade:",grade)

```

Question 1 : Find the sum of first 10 natural numbers

```python
sum = 0
for i in range(1,11) :
  sum = sum + i
print("sum of first 10 natural numebrs :", sum)
```

Question 2 : Find factorial of a number

```python
n = int(input("enter a number :"))
fact = 1
for i in range(1,n+1) :
  fact = fact * i
print("factorial of n is ",fact)

```

Question 3 : Print fibonacci series

```python
n = int(input("enter length of the series :"))
a, b  = 0 , 1
print(1, end=" ")
for i in range(n):
  print(a+b, end = " ")
  a,b = b , a+b
```

Question 4 : Find Largest among three numbers

```python
a = int(input("enter first number :"))
b = int(input("enter second number :"))
c = int(input("enter third number :"))
if a>b and a>c :
  print(a,"is largest")
elif b>a and b>c :
  print(b,"is largest")
else :
  print(c,"is largest")
```

Question 5 : Create Student Result System
- input student details
- input marks
- Calculate percentage
- Display grade


```python
name = input("enter name :")
roll_number = input("enter roll number :")
math_marks = int(input("enter math marks :"))
science_marks = int(input("enter science marks :"))
english_marks = int(input("enter english marks :"))
sst_marks = int(input("enter sst marks :"))
hindi_marks = int(input("enter hindi marks :"))
total_marks = math_marks + science_marks + english_marks + sst_marks + hindi_marks
percentage = (total_marks / 500) * 100

if percentage >= 90 :
  grade = "A"
elif percentage >= 80 :
  grade = "B"
elif percentage >= 70 :
  grade = "C"
elif percentage >= 60 :
  grade = "D"
elif percentage >= 50 :
  grade = "E"
else :
  grade = "F"

print("name:",name )
print("roll number:",roll_number)
print("math marks:",math_marks)
print("science marks:",science_marks)
print("english marks:",english_marks)
print("sst marks:",sst_marks)
print("hindi marks:",hindi_marks)
print("total marks:",total_marks)
print("percentage:",percentage)
print("grade:",grade)

```

```python
# function to print first 10 natural numbers
def natural_num():
    for i in range (1,11):
        print(i)
natural_num()

#function to calculate sum of first N natural numbers
def sum_natural(n):
    sum = 0
    for i in range (1,n+1):
        sum = sum + i
    return sum
n = int(input("Enter a number: "))
print("Sum of first", n, "natural numbers is:", sum_natural(n))

#function to reverse a number
def reverse_num(num):
   rev = int(str(num)[::-1])
   return rev
num = int(input("Enter a number: "))
print("Reverse of the number is:", reverse_num(num))

#function to count digits in a number
def count_digits(a):
    count = 0
    while a > 0:
        count = count + 1
        a = a // 10
    return count
a = int(input("Enter a number to count digits: "))
print("Number of digits in the number is:", count_digits(a))

#function to check palindrome number
def palindrome(b):
    rev=int(str(b)[::-1])
    if b == rev:
        print("The number is a palindrome.")
    else:
        print("The number is not a palindrome.")
b = int(input("Enter a number to check palindrome: "))
palindrome(b)

#function to generate Fibonacci series 
def fibonacci(N):
    a, b = 0, 1
    for i in range(N):
        print(a, end=' ')
        a = b
        b = a + b
N= int(input("Enter the number of terms in Fibonacci series: ")
fibonacci(N)

# Calculator using Functions
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Division by zero is not allowed!"
    return x / y

# Display Menu
print("Calculator")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

# User selects operation
choice = int(input("Enter your choice (1-4): "))

# Input numbers
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform calculation
if choice == 1:
    result = add(num1, num2)
    print("Result =", result)

elif choice == 2:
    result = subtract(num1, num2)
    print("Result =", result)

elif choice == 3:
    result = multiply(num1, num2)
    print("Result =", result)

elif choice == 4:
    result = divide(num1, num2)
    print("Result =", result)

else:
    print("Invalid Choice!")

# Create a text file and store student details
name = "MANAVI TULASYAN"
roll_no = "0720118"
course = "B.Tech"
# Open file in write mode
file = open("student_details.txt", "w")
file.write("Name: " + name )
file.write("Roll No: " + roll_no )
file.write("Course: " + course )
file.close()
print("Student details saved successfully in 'student_details.txt'")

#read data from a file
file = open("student_details.txt", "r")
content = file.read()
print("Student Details:")
print(content)

#handle division by zero using exception handling
def safe_divide(x, y):
    try:
        result = x / y
        return result
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed!"
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
result = safe_divide(num1, num2)
print("Result =", result)

#create a student class with name and marks
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks) 
student1 = Student("MANAVI TULASYAN", 85)
student1.display()

```

```python
# Student Report Program
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")
SS = float(input("Enter marks in SS: "))
FES= float(input("Enter marks in FES: "))
NAS = float(input("Enter marks in NAS: "))
IDS = float(input("Enter marks in IDS: "))
AP = float(input("Enter marks in AP: "))
# Calculating total marks
total = SS+FES+NAS+IDS+AP
# Calculating percentage
percentage = (total / 500) * 100
print("----- STUDENT REPORT -----")
print("Name:", name)
print("Roll Number:", roll_no)
print("Total Marks:", total, "/ 500")
print("Percentage:", percentage, "%")
```

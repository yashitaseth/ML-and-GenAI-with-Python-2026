1. Find area of rectangle

```python
len = float(input("Enter length: "))
b = float(input("Enter breadth: "))

area = len * b

print("Area of Rectangle =", area)
```

2) Find simple interest

```python
p = float(input("Enter Principal Amount: "))
r = float(input("Enter Rate of Interest: "))
t = float(input("Enter Time: "))

simpleint = (p * r * t) / 100

print("Simple Interest =", simpleint)
```

3) Convert temperature from Celsius to Fahrenheit

```python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32

print("Temperature in Fahrenheit =", fahrenheit)
```

4. Calculate average of 3 numbers.

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

average = (num1 + num2 + num3) / 3

print("Average =", average)
```

5. Find square and cube of a number.

```python
num = int(input("Enter a number: "))

square = num ** 2
cube = num ** 3

print("Square =", square)
print("Cube =", cube)
```

6. Swap two numbers without third variable.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a = a + b
b = a - b
a = a - b

print("After Swapping:")
print("a =", a)
print("b =", b)
```

7. Create a Student Report Program that take student details using input(), Store marks in variables, Calculate total and percentage , Use comments, Use proper indentation

```python
name = input("Enter Student Name: ")
roll_no = input("Enter Roll Number: ")

mark1 = float(input("Enter marks of Subject 1: "))
mark2 = float(input("Enter marks of Subject 2: "))
mark3 = float(input("Enter marks of Subject 3: "))

total = mark1 + mark2 + mark3
percentage = total / 3

print("-----76STUDENT REPORT -----")
print("Name :", name)
print("Roll No :", roll_no)
print("Marks:")
print("Subject 1 =", mark1)
print("Subject 2 =", mark2)
print("Subject 3 =", mark3)
print("Total Marks =", total)
print("Percentage =", percentage, "%")
```

```python
#To find simple interest
p = int(input("Enter principal amount:"))
r = int(input("Enter the rate of interest in %:"))
t = int(input("Enter time in years:"))
simple_interest=(p*r*t)/100
print("The simple interest is:",simple_interest)
```

```python
#4 count digits in a number
def digits(x):
    return len(str(x))
digits(245465)
```

```python
#5 check palindrome number
def check_pallindrome(x):
  return str(x)==str(x)[::-1]
check_pallindrome(456654)
```

```python
#6 fibonacci series
def fib(n):
    def f(x):
        if x==0:
            return 0
        elif x==1:
            return 1
        return f(x-1)+f(x-2)
    for i in range(n):
        print(f(i), end=" ")
fib(8)
```

```python
# Assignment-1
#1 Area of Rectangle
# Let 'l' and 'b' be the length and breadth of the rectangle respectively
l=float(input("Enter the length of the rectangle:"))
b=float(input("Enter the breadth of the rectange"))
print("Area of the rectangle is ",l*b,"square units")

```

```python
#2 Simple Interest
# Let p,r and t denote Principal amount, Rate of interest and Time Duration respectively
p=int(input("Enter the Principal Amount in Rupees:"))
r=float(input("Enter the Rate of Interest in % p.a.:"))
t=int(input("Enter the Time in years:"))
print("Simple Interest=",(p*r*t)/100,"Rupees")
```

```python
#3 Celsius to Fahrenheit temperature conversion
c=float(input("Enter temperature in degree Celsius:"))
f=(1.8*c)+32
print("Corresponding Fahrenheit tempertature is",f,"degree F")
```

```python
#4 Average of 3 numbers
a=float(input("Enter the 1st number:"))
b=float(input("Enter the 2nd number:"))
c=float(input("Enter the 3rd number:"))
d=(a+b+c)/3
print("Average of the given numbers is",d)
```

```python
#5 Square and Cube of a number
x=float(input("Enter the number:"))
sq=x**2
cb=x**3
print("The square and cube of the given numbers are",sq,"and",cb,"respectively.")
```

```python
#6 Swap two variables without using a third one
a = int(input("Enter 1st number:"))
b = int(input("Enter 2nd number:"))
print("Original numbers: a=",a,"and b=",b)
a, b = b, a
print("Swapped numbers: a=",a,"and b=",b)
```
